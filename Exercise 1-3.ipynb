```python
SEED_NUM = 9                          
STUDENT_MAJOR = "BSECE"              
FAVORITE_ARTIST = "Toneejay"       
LOCAL_HAZARD = "Traffic"             
CONTROL_NUM = max(3, SEED_NUM)
# =======================================

print(f"CONTROL_NUM: {CONTROL_NUM}")
print(f"STUDENT_MAJOR: {STUDENT_MAJOR}")
print(f"FAVORITE_ARTIST: {FAVORITE_ARTIST}")
print(f"LOCAL_HAZARD: {LOCAL_HAZARD}")
```

```python
from collections import defaultdict, Counter

# Initialize stream
stream = [
    ("TUP Manila", "BSECE"),
    ("TUP Taguig", f"BSME_v{CONTROL_NUM}"),
    ("TUP Manila", f"BSME_v{CONTROL_NUM}"),
    ("TUP Manila", "BSECE"),
    ("TUP Visayas", STUDENT_MAJOR),
    ("TUP Taguig", "BSECE"),
    ("TUP Manila", "BSECE")
]

# Data Expansion: add exactly (CONTROL_NUM + 3) custom tuples
custom_apps = [
    ("TUP Manila", "BSIT"),
    ("TUP Manila", "BSEE"),
    ("TUP Taguig", "BSIT"),
    ("TUP Visayas", "BSECE"),
    ("TUP Manila", "BSCE"),
    ("TUP Cavite", "BSME"),
    ("TUP Manila", "BSECE"),
]
stream.extend(custom_apps[:CONTROL_NUM + 3])

# Algorithm Phase 1 - Aggregation
campus_dict = defaultdict(list)
for campus, program in stream:
    campus_dict[campus].append(program)

# Algorithm Phase 2 - Frequency Analysis
campus_counters = {}
for campus, programs in campus_dict.items():
    campus_counters[campus] = Counter(programs)

# Extract top program at TUP Manila
top_program, top_freq = campus_counters["TUP Manila"].most_common(1)[0]

# Print Assessment Table outputs
print(f"CONTROL_NUM Used: {CONTROL_NUM}")
print(f"STUDENT_MAJOR Used: {STUDENT_MAJOR}")
print(f"Expanded Stream Length (After adding data): {len(stream)}")
print(f"Total Applications Processed for TUP Manila: {sum(campus_counters['TUP Manila'].values())}")
print(f"Top Requested Program at TUP Manila: {top_program}")
print(f"Exact Frequency of Top Program: {top_freq}")
```

```python
# Initialize Festival Lineup as a Set
festival = {
    "Ben&Ben", "SB19", "Bini", "Eraserheads",
    FAVORITE_ARTIST, "Zild", f"Indie Artist {CONTROL_NUM}"
}

# Initialize User Profiles
user_a = {"Ben&Ben", "Bini", "Maki", "Dionela", FAVORITE_ARTIST}
user_b = {"SB19", "Eraserheads", "Zild", f"Indie Artist {CONTROL_NUM}", "Parokya ni Edgar"}

# Data Expansion: add exactly CONTROL_NUM new artists to festival
new_artists = {f"Artist_{i}" for i in range(1, CONTROL_NUM + 1)}
festival.update(new_artists)

# Algorithm
# 1. Jaccard Similarity for User A
intersection_a = user_a & festival
union_a = user_a | festival
jaccard_a = (len(intersection_a) / len(union_a)) * 100

# 2. Jaccard Similarity for User B
intersection_b = user_b & festival
union_b = user_b | festival
jaccard_b = (len(intersection_b) / len(union_b)) * 100

# 3. User A Dealbreakers (User A listens to but NOT in festival)
dealbreakers_a = user_a - festival

# Print Assessment Table outputs
print(f"Total Artists in Expanded Festival: {len(festival)}")
print(f"Length of Intersection (User A & Festival): {len(intersection_a)}")
print(f"Length of Union (User A & Festival): {len(union_a)}")
print(f"User A Jaccard Similarity (%): {jaccard_a:.2f}%")
print(f"User B Jaccard Similarity (%): {jaccard_b:.2f}%")
print(f"User A Dealbreaker Artists (Set Output): {dealbreakers_a}")
```

```python
from collections import deque, Counter

# Initialize deque with maxlen = 5
alert_deque = deque(maxlen=5)

# Base burst list
burst = [
    (CONTROL_NUM,     "WEATHER"),
    (CONTROL_NUM + 1, "TRAFFIC"),
    (CONTROL_NUM + 2, "WEATHER"),
    (CONTROL_NUM + 3, "WEATHER"),
    (CONTROL_NUM + 4, LOCAL_HAZARD),
    (CONTROL_NUM + 5, "WEATHER"),
    (CONTROL_NUM + 6, "TRAFFIC")
]

# Data Expansion: append exactly (CONTROL_NUM + 2) custom alerts
# Timestamps continue from where burst left off
last_ts = CONTROL_NUM + 6
custom_alerts = []
alert_types = ["TRAFFIC", "WEATHER", LOCAL_HAZARD, "WEATHER", "TRAFFIC", "WEATHER", LOCAL_HAZARD]
for i in range(CONTROL_NUM + 2):
    last_ts += 1
    custom_alerts.append((last_ts, alert_types[i % len(alert_types)]))

burst.extend(custom_alerts)

# Algorithm — Sliding Window
rejected = []
for alert in burst:
    timestamp, category = alert

    # Extract current categories in deque
    current_categories = [item[1] for item in alert_deque]

    # Count frequencies
    category_count = Counter(current_categories)

    # Spam Rule: reject if category already appears 2 or more times
    if category_count[category] >= 2:
        rejected.append(alert)
    else:
        alert_deque.append(alert)

# Sum all Timestamp values in final deque
timestamp_sum = sum(item[0] for item in alert_deque)

# Print Assessment Table outputs
print(f"Final Deque State: {list(alert_deque)}")
print(f"Total Alerts Rejected (Spam): {len(rejected)}")
print(f"Sum of Timestamps in Final Deque: {timestamp_sum}")
print(f"Alerts in Final Deque: {len(alert_deque)}")
```
